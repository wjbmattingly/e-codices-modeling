```python
import pandas as pd
import glob
import json
```

```python
files = glob.glob("json_data/*.json")
len(files)
```

```python
df = pd.read_csv("data/e_codices_database.csv")
df.head(1)
```

```python
all_data = []
for idx, row in df.iterrows():
    url_parts = row.url.split("/")
    thumb_url = f"https://www.e-codices.unifr.ch/en/thumbs/{url_parts[-2]}/{url_parts[-1]}"
    json_file = f"json_data/{url_parts[-2]}_{url_parts[-1]}.json"
    try:
        with open(json_file, "r", encoding="utf-8") as f:
            data = json.load(f)
        for item in data["canvases"]:
            item["shelfmark"] = row.shelfmark
            all_data.append(item)
    except:
        Exception
```

```python
index = pd.DataFrame(all_data)
index.to_csv("index.csv", index=False)
```
