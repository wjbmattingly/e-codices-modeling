```python
import json
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def  get_mss(url):
    s = requests.get(url)
    soup = BeautifulSoup(s.content)
    divs = soup.find_all("div", {"class": "search-result"})
    print (len(divs))
    links = []
    for div in divs:
        shelfmark = div.find("div", {"class": "collection-shelfmark"}).text.strip()
        headline = div.find("div", {"class": "document-headline"}).text.strip()
        summary = div.find("p", {"class": "document-summary-search"}).text.strip()
        link = div.find("a")["href"]
        image_soup = div.find("a").find("iiif-image")
        image_server_base_url = image_soup['image-server-base-url']
        image_rel = image_soup['image-rel']
        links.append({"shelfmark": shelfmark, "headline": headline, "summary": summary,
                      "url": link,
                      "image_soup": image_soup, "image_server_base_url":image_server_base_url,
                        "image_rel": image_rel})
    df = pd.DataFrame(links)
    return df
df = get_mss(url="https://www.e-codices.unifr.ch/en/search/?iCurrentPage=1&sQueryString=&sSearchField=fullText&iResultsPerPage=10000&iCurrentPage=1&sSortField=score")
df
```

```python
df.to_csv("data/e_codices_database.csv", index=False)
```
