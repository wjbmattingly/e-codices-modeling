```python
import pandas as pd
import requests
from PIL import Image
from io import BytesIO
from bs4 import BeautifulSoup
import json
import time
import random
```

```python
df = pd.read_csv("data/e_codices_database.csv")
```

```python
df.head(1)
```

```python
rels = df.image_rel.tolist()
headlines = df.headline.tolist()
shelfmarks = df.shelfmark.tolist()
urls = df.url.tolist()
urls[1]
```

```python
df.loc[1].url
```

```python
i=0
for original_url, rel, headline, shelfmark in zip(urls, rels, headlines, shelfmarks):
    shelfmark = shelfmark.replace(" ", "_").lower()
    url_parts = original_url.split("/")
    thumb_url = f"https://www.e-codices.unifr.ch/en/thumbs/{url_parts[-2]}/{url_parts[-1]}"
    print(i, shelfmark, thumb_url)
    time.sleep(random.randint(5, 30))
    s = requests.get(thumb_url)
    soup = BeautifulSoup(s.content)
    viewer = soup.find("manuscript-viewer")
    viewer_data = viewer[":viewer-data"]
    json_data = json.loads(viewer_data)
    json_file = f"json_data/{url_parts[-2]}_{url_parts[-1]}.json"
    with open(json_file, "w", encoding="utf-8") as f:
        json.dump(json_data, f, indent=4)
    i=i+1
```
