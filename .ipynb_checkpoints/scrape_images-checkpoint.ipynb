```python
import pandas as pd
import requests
from PIL import Image
from io import BytesIO
```

```python
df = pd.read_csv("data/e_codices_database.csv")
```

```python
df
```

```python
rels = df.image_rel.tolist()
headlines = df.headline.tolist()
shelfmarks = df.shelfmark.tolist()
```

```python
for rel, headline, shelfmark in zip(rels[1:2], headlines[1:2], shelfmarks[1:2]):
    url = rel.replace("pct:25", ",300")
    print (url)
    print (headline)
    shelfmark = shelfmark.replace(" ", "_").lower()
    print (shelfmark)
    for i in range(1, 1169):
        image_url = url.replace("000h", f"{i:03}")
        print (image_url)
        response = requests.get(url)
        img = Image.open(BytesIO(response.content))
        img.save(f"images/kba/kba-0016-1/{i:04}.jpg")
```
